### Data Cleaning, Feature Engineering and Analysis

```python
profiles_df['is_verified'].unique()
```

```python
profiles_df['is_verified'].replace({False:0, True: 1}, inplace=True)
profiles_df['is_verified'].unique()
```

```python
# Function to group influencers
def influencer_category(infcat):
    if infcat < 10000:
        return 'normal_user'
    elif infcat < 500000:
        return 'micro_influencer'
    else:
        return 'macro_influencer'

profiles_df['Inlfuencer Category'] = profiles_df['followers'].apply(influencer_category)
profiles_df.head(5) 
```

```python
profiles_df['Inlfuencer Category'].value_counts()
```

```python
len(profiles_df)
```

```python
# Removing all the unnecessary columns for segmentation and analysis
cluster_df = profiles_df[['awg_engagement_rate', 'comment_engagement_rate', 'like_engagement_rate','is_verified','followers','following','likes','videos_count','id']]
cluster_df.head(5)
```

### K-Means Clustering model

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample engagement data (replace with your dataset)
engagement_data = cluster_df

# Convert to DataFrame
data = pd.DataFrame(engagement_data[['followers','following']])

# Data preprocessing by normalizing the dataset
scaler = StandardScaler()
scaled = scaler.fit_transform(data)

from scipy import stats

# Assuming 'data' is your DataFrame and 'threshold' is the Z-score threshold
z_scores = stats.zscore(scaled)
X_scaled = data[(np.abs(z_scores) < 2).all(axis=1)]


pca = PCA(n_components=2)  # Reduce to 2 principal components for visualization
pca_result = pca.fit_transform(X_scaled)

# Add PCA results to DataFrame
X_scaled['PCA1'] = pca_result[:, 0]
X_scaled['PCA2'] = pca_result[:, 1]

# Select PCA components for clustering
features_for_clustering = ['PCA1', 'PCA2']

# Perform k-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Example: 5 clusters
clusters = kmeans.fit_predict(X_scaled[features_for_clustering])

# Add cluster labels to DataFrame
X_scaled['cluster'] = clusters

# Visualize the clusters
for cluster in range(3):
    cluster_data = X_scaled[X_scaled['cluster'] == cluster]
    plt.scatter(cluster_data['followers'], cluster_data['following'], label=f'Cluster {cluster}')
plt.xlabel('Followers')
plt.ylabel('Following')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

```python
print('Min Values:')
min_values_per_cluster = X_scaled.groupby('cluster')['followers'].min()
print(min_values_per_cluster)
print('Max Values:')
min_values_per_cluster = X_scaled.groupby('cluster')['followers'].max()
print(min_values_per_cluster)
print('Value counts:')
print(X_scaled['cluster'].value_counts())
```

```python
# Adding all the columns back to perform analysis on reach and engagement per cluster
missing_columns = [col for col in cluster_df.columns if col not in X_scaled.columns]

# Copy missing columns from df1 to df2
for col in missing_columns:
    X_scaled[col] = cluster_df[col].copy()
```

```python
print('Reach:')
reach = X_scaled.groupby('cluster')['likes'].mean()
print(reach)
```

```python
print('Reach:')
reach = X_scaled.groupby('cluster')['awg_engagement_rate'].mean()
print(reach)
```

### NLP - Topic Modelling on Reviews

```python
reviews_df = pd.read_csv("tiktok_app_reviews.csv")
reviews_df.head(5)
print(len(reviews_df))
```

```python
from gensim import corpora
from gensim.models import LdaModel
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Sample TikTok reviews (replace with your dataset)
reviews = reviews_df['review_text']

# Preprocessing
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
cleaned_reviews = []
for review in reviews:
    review = str(review)
    tokens = word_tokenize(review.lower())  # Tokenization and lowercasing
    tokens = [token for token in tokens if token not in stop_words and token not in punctuation]  # Remove stopwords and punctuation
    cleaned_reviews.append(tokens)

# Create dictionary and corpus
dictionary = corpora.Dictionary(cleaned_reviews)
corpus = [dictionary.doc2bow(review) for review in cleaned_reviews]

# Train LDA model
lda_model = LdaModel(corpus, num_topics=3, id2word=dictionary, passes=10)

# Print topics
print("Topics:")
for topic in lda_model.print_topics():
    print(topic)
```
